```python
import torchaudio
import matplotlib.pyplot as plt
```

```python
bed_sample, bed_sampling_rate = torchaudio.load("train/audio/bed/0a7c2a8d_nohash_0.wav")
nine_sample, nine_sampling_rate = torchaudio.load("train/audio/nine/0a7c2a8d_nohash_0.wav")
right_sample, right_sampling_rate = torchaudio.load("train/audio/right/0a7c2a8d_nohash_0.wav")
```

### Plot sample 1D

```python
plt.figure(figsize=(20, 10))

plt.subplot(3, 1, 1)
plt.plot(bed_sample.t().numpy(), linewidth=0.5, color='orange')
plt.title("'bed' sample waveform")
plt.xlabel("Sample number")
plt.ylabel("Aplitude")
plt.ylim(-0.4, 0.4)

plt.subplot(3, 1, 2)
plt.plot(nine_sample.t().numpy(), linewidth=0.5, color='orange')
plt.title("'nine' sample waveform")
plt.xlabel("Sample number")
plt.ylabel("Aplitude")
plt.ylim(-0.4, 0.4)

plt.subplot(3, 1, 3)
plt.plot(right_sample.t().numpy(), linewidth=0.5, color='orange')
plt.title("'right' sample waveform")
plt.xlabel("Sample number")
plt.ylabel("Aplitude")
plt.ylim(-0.4, 0.4)

plt.subplots_adjust(hspace=0.5)
plt.show()
```
